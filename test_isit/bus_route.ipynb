```python
import pandas as pd
from geopy.distance import geodesic

# Load necessary GTFS files
routes = pd.read_csv("C:/Users/z048523/myenv/routes.csv")
trips = pd.read_csv("C:/Users/z048523/myenv/trips.csv")
stop_times = pd.read_csv("C:/Users/z048523/myenv/stop_times.csv")
stops = pd.read_csv("C:/Users/z048523/myenv/stops.csv")
```

```python
import pandas as pd
from datetime import datetime



# Filter trips for route 17
route_17_trips = trips[trips['route_id'].astype(str) == "17"]
trip_ids_17 = route_17_trips['trip_id'].unique()

# Get stop times for route 17 trips
route_17_stop_times = stop_times[stop_times['trip_id'].isin(trip_ids_17)].copy()

# Merge with stop data to get stop names or coordinates if needed
route_17_stop_times = route_17_stop_times.merge(stops[['stop_id', 'stop_lat', 'stop_lon']], on='stop_id', how='left')

# Convert departure time to datetime (removing invalid values like 24:xx)
def parse_time(t):
    try:
        return datetime.strptime(t, "%H:%M:%S").time()
    except:
        return None

route_17_stop_times['departure_time'] = route_17_stop_times['departure_time'].apply(parse_time)
route_17_stop_times = route_17_stop_times.dropna(subset=['departure_time'])

# Count how many departures per hour
route_17_stop_times['hour'] = route_17_stop_times['departure_time'].apply(lambda t: t.hour)
frequency_by_hour = route_17_stop_times.groupby('hour').size()




```

```python
# Unique trip IDs for each hour based on their first stop time
first_stops = route_17_stop_times.sort_values(['trip_id', 'stop_sequence']).groupby('trip_id').first().reset_index()

# Extract the hour of departure
first_stops['hour'] = first_stops['departure_time'].apply(lambda t: t.hour)

# Trips per hour
trips_per_hour = first_stops.groupby('hour').size()

# More accurate average
avg_frequency = trips_per_hour.mean()

print("✅ Unique trips on route 17:", len(first_stops))
print("✅ Average frequency (buses/hour):", round(avg_frequency, 2))
print("\n📊 Hourly breakdown:\n", trips_per_hour)
```
